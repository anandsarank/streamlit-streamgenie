```python
import pandas as pd
import numpy as np

netflix = pd.read_csv("/Users/anandsaran/Documents/Code/Projects/Streamlit Applications/StreamGuru/data/netflix_titles.csv")
prime = pd.read_csv("/Users/anandsaran/Documents/Code/Projects/Streamlit Applications/StreamGuru/data/amazon_prime_titles.csv")
disney = pd.read_csv("/Users/anandsaran/Documents/Code/Projects/Streamlit Applications/StreamGuru/data/disney_plus_titles.csv")
```

```python
print(netflix.info())
```

```python
netflix['platform'] = 'Netflix'
prime['platform'] = 'Amazon Prime Video'
disney['platform'] = 'Disney Plus'
netflix['show_id'] = "n_"+netflix['show_id']
prime['show_id'] = "p_"+prime['show_id']
disney['show_id'] = "d_"+disney['show_id']

ott_data = pd.DataFrame(columns = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'listed_in', 'description', 'platform'])

ott_data = ott_data.append(netflix)
ott_data = ott_data.append(prime)
ott_data = ott_data.append(disney)
ott_data.info()
```
